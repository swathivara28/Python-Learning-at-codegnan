```python
# Built in Functions: The functions definition is provided by python,those function called as Built in functions
# abs()/absolute(): It return thr absolute value
print(abs(20-10))
print(abs(10-20))
```

```python
# round(val,pos): This function rounds the float value to particular position
print(round(2.2589,2))
print(round(2.2539,2))
print(round(2.2539))
print(int(2.2539))
```

```python
# ord(char):It returns ACSII Value of character
# char(ascii_val):It return the character for specified ASCII value
print(ord("A"))
print(chr(65))
```

```python
# sorted(iterable,reverse=False,key=True):It is a built in function but,[sort is list function] It returns new sorted iterable
# sort the fruits based on the length of fruit names
fruits = ['Apple','Banana','Kiwi','Mango','Water melon']
res = sorted(fruits)
print(res)
print(sorted(fruits,reverse=True))
print(sorted(fruits,reverse=True,key =len)) # it sorts the list based on length of each
```

```python
# map(function,iterable)
# find the square of all list of integers using map function
def square(x):
    return x**2
arr = [1,2,3,4,5]
res = list(map(square,arr))
print(res)
```

```python
#filter(function,iterable) : It filters the iterable,if the function returns non-none value,those values will added to result
# find the even numbers from list of numbers
def check_even(x):
    if x % 2==0:
        return x
lst = [1,2,3,4,5,6]
res = list(filter(check_even,lst))
print(res)
```

```python
#filter(function,iterable)
# find the even numbers from list of numbers
def check_even(x):
    if x % 2==0:
        return x
lst = [1,2,3,4,5,6]
res = list(map(check_even,lst))
print(res)
```

```python
# find the three multipliers from list of integers
#filter(function,iterable)
# find the even numbers from list of numbers
arr = [1,24,54,22,44,51,33,15]
def Threes_multipliers(x):
    if x % 3==0:
        return 1
res = list(filter(Threes_multipliers,arr))
print(res)
```

```python
# enumerate(iterable): This function returns the list of tuple of index and item for iterable items
fruits = ['Mango','Apple','banana','Kivi','Water melon']
print(enumerate(fruits))
print(list(enumerate(fruits)))
for ind, val in enumerate(fruits):
    print(f"{ind}-{val} and length is {len(val)}")
```

```python
# Tuple Unpacking
t =(1,2)
a,b = t
print(a,b)
```

```python
# Zip(iterable,iterable2,....): It Returns all iterables as single iterable of tuples
rool_no = [101,102,103,104,105,]
names = ['Ravi','Raju','siva','Rani','Gopi']
percentage = [90,96.8,67,65,99.99999]
print(list(zip(rool_no,names,percentage)))
```

```python
# 1.num can be modified into binary format
num = int(input())
res = " "
while num> 0:
    bit = num % 2
    num = num// 2
    res = str(bit)+res
print(res)
```

```python
# any(): This function returns True when any one iterable items return True
# all(): Return True if bool(x) is True for all values x in the iterable
arr = [1,2,3,4,0]
print(all(arr))
print(any(arr))
```

```python
percentage = [34,35,54,65,67]
for num in percentage:
    if num < 35:
        print("All Students not Passed")
        break
    else:
        print("All students pass")
```

```python
percentage = [34,35,54,65,67]
lst = []
for num in percentage:
    if num < 35:
        lst.append(False)
    else:
        lst.append(True)
print(lst)
if all(lst):
    print("All students pass")
else:
    print("All Students not Passed")
```
