```python
# List comprehinsion 
# Syntax: [exp for iterator in iterable if condition]
# return 1-n numbers as list using list comprehinsion
n = int(input())
res = [x for x in range(1,n+1)]
print(res)
```

```python
n = int(input())
l = []
for i in range(1, n + 1):
    l.append(i)
print(l)
```

```python
# find the first  n even number
n = int(input())
res = [2*x for x in range(n)]
print(res)
```

```python
# find the first  n even number
# Another way to finf n even number
n = int(input())
res = [x for x in range(0,2*n,2) if x%2==0]
print(res)
```

```python
# return list even or odd,for list of integers
arr = [20,34,33,55,23,54]
res = ["odd" if x&1 else "even" for x in arr]
print(res)
```

```python
# return list even or odd,for list of integers
arr = [20,34,33,55,23,54]
res = [x +0.5 if x&1 else x+1 for x in arr]
print(res)
```

```python
#find one dice combinations
print([x for x in range(1,7)])
```

```python
# find two dice combinations
print([(i,j) for i in range(1,7) for j in range(1,7)]) 
```

```python
# find two dices combinations if any dice face is 4
print([(i,j) for i in range(1,7) for j in range(1,7) if i==4 or j==4]) 
```
