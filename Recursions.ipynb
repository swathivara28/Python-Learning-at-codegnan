```python
# Factorial of a numbers using Recursion function
def factorial(n):
    if n == 1 or n==0:
        return 1
    return n*factorial(n-1)
factorial(6)
```

```python
# Digit to binary number conversion using Recursion function
def binary(n):
    if n==0:
        return "0"
    if n==1:
        return "1"
    return binary(n//2)+str(n%2)
binary(4)
```

```python
# square of two numbers using lambda
square = lambda x:x**2
print(square(4))
```

```python
# sum of two numbers using lambda
add = lambda x,y:x+y
print(add(4,3))
```

```python
#find even numbers in list using lambda and filter()
arr = [1,2,3,4,5]
res = list(filter(lambda x:x%2==0,arr))
print(res)
```

```python
# single line if-else syntax:
# True_exp if condition else False_exp
# find the weather number is even or odd using lambda
n = int(input())
even_or_odd = lambda x:f"{x} odd" if x%2 else f"{x} even"
print(even_or_odd(n))
```

```python
# sum of natural numbers using Recursion
def sum_of_natural_numbers(n):
    if n==1:
        return 1
    return n+sum_of_natural_numbers(n-1)
sum_of_natural_numbers(8)
```

```python
# find fabanoic sereies using Recursion
def fabanoic(n):
    if n==1 or n==0:
        return 1
    return fabanoic(n-1)+fabanoic(n-2)
fabanoic(6)
```
