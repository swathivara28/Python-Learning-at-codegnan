```python
# Types of Arguments
# 1. Positional Arguments:The arguments are assigns to parameters as per the positional order
def sub(x:int,y:int)->int:
    c = x-y
    return c
# function calling block
a,b = map(int,input().split())
res = sub(a,b)
print(res)
```

```python
# 2 numbers sum 
def add(x,y,z):
    c = x+y+z
    return c
# function calling block
a,b,c = 10,20,30
res = add(a,b,c)
print(res)
```

```python
# 3.Default Arguments: The parameters defined with some default values,
#If you pass the values for deafult parameters otherwise , it will assign default values
def add(a,b,c = 0,d= 0):
    res = a + b + c + d
    return res
print(add(10,20))
print(add(10,20,5))
print(add(10,20,2,1))
print(add(10))
```

```python
# 4. Variable Length of positional argument
def add(a, b = 0 , *args):
    print(type(args))
    print(args)
print(add(1,2,3,4,5,6,7,8,9))
```

```python
# 4. Variable Length of positional argument
def add(a, b = 0 , *args):
    print(type(args))
    print(args)
    s = 0
    for num in args:
        s += num
    return s + a + b
print(add(1))
#print(add(1,2, b =3))
print(add(1,2,3,5,5))
print(add(1,2,3,4,5,6,7,8,9))
```

```python
def add(a,b,c = 0,d= 0):
    res = a * b *c * d
    return res
print(add(10,20))
print(add(10,20,5))
print(add(10,20,2,1))
```
