```python
X = input().strip()
if '0' <= X <= '9':
    print("IS DIGIT")
elif 'A' <= X <= 'Z':
    print("ALPHA")
    print("IS CAPITAL")
elif 'a' <= X <= 'z':
    print("ALPHA")
    print("IS SMALL")
```

```python
X = input().strip()

if 'a' <= X <= 'z':
    print(X.upper())
else:
    print(X.lower())
```

```python
s = input().strip()

if '+' in s:
    A, B = map(int, s.split('+'))
    print(A + B)
elif '-' in s:
    A, B = map(int, s.split('-'))
    print(A - B)
elif '*' in s:
    A, B = map(int, s.split('*'))
    print(A * B)
elif '/' in s:
    A, B = map(int, s.split('/'))
    print(A // B) # // for integer division
```

```python
X = input().strip()
firstDigit = int(X[0]) # take first character

if firstDigit % 2 == 0:
    print("EVEN")
else:
    print("ODD")
```

```python
X, Y = map(float, input().split())

if X == 0 and Y == 0:
    print("Origem")
elif X == 0:
    print("Eixo Y")
elif Y == 0:
    print("Eixo X")
elif X > 0 and Y > 0:
    print("Q1")
elif X < 0 and Y > 0:
    print("Q2")
elif X < 0 and Y < 0:
    print("Q3")
elif X > 0 and Y < 0:
    print("Q4")
```

```python
N = int(input())

years = N // 365
N = N % 365

months = N // 30
days = N % 30

print(f"{years} years")
print(f"{months} months")
print(f"{days} days")
```

```python
X = float(input())

if 0 <= X <= 25:
    print("Interval [0,25]")
elif 25 < X <= 50:
    print("Interval (25,50]")
elif 50 < X <= 75:
    print("Interval (50,75]")
elif 75 < X <= 100:
    print("Interval (75,100]")
else:
    print("Out of Intervals")
```

```python
N = float(input())
int_part = int(N)
dec_part = N - int_part
if dec_part == 0:
    print("int", int_part)
else:
    # decimal part ni 3 digits varaku print cheyali
    print("float", int_part, f"{dec_part:.3f}"[1:]) 
```

```python
A, S, B = input().split()
A, B = int(A), int(B)

if (S == '<' and A < B) or (S == '>' and A > B) or (S == '=' and A == B):
    print("Right")
else:
    print("Wrong")
```
