```python
# count even and odd numbers in given list of numbers
# separete the even and odd number in given list 
# List = [1,2,3,4,5,6] o/p : even[2,4,6] , odd:[1,3,5]
```

```python
arr = list(map(int,input().split()))
even = odd =0
for num in arr:
    if num % 2 == 0:
       even += 1
    else:
        odd += 1
print(even,odd)
```

```python
arr = list(map(int,input().split()))
even = []
odd = []
for num in arr:
    if num % 2 == 0:
       even.append(num)
    else:
        odd.append(num)
print(even,odd)
```
