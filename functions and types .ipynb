```python
# 4. with return and with return
# sum of two numbers
#function definition
def add(x:int, y:int)->int:
    c = x+y
    return c
# function calling
a,b = 5,6
res = add(a,b)
print(res)
print(add(1,2))
print(add(3,4))      
```

```python
# Find the weather the number is prime number or not using function
def prime_check(n:int)->bool:
    if n< 2:
        return False
    for i in range(2,n//2+1):
        if n % i==0:
            return False
    return True
num = int(input())
print(prime_check(num))
```

```python
# Find the weather the number is prime number or not using function
def prime_check(n:int)->bool:
    for i in range(2,n//2+1):
        if n % i==0:
            return False
    return True
for j in range(1,100):
    if prime_check(j):
        print(j,end=" ")
```
