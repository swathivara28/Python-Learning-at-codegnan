```python
# Enclosing scope
def outer(x):
    y = 10 # Enclosing Scope
    def inner():
        a = 10 # Local Scope
        z = y + a
        print(f"This is local scope:{z}")
    inner()
    y = y + x
    print(f"This is enclosing scope:{y}")
n = 5
outer(n)
```

```python
# Enclosing scope
def outer(x):
    x = 10 # Enclosing Scope
    def inner():
        x = 20 # Local Scope
        z = x + x
        print(f"This is local scope:{z}")
    inner()
    y = x + x
    print(f"This is enclosing scope:{y}")
n = 5
outer(n)
```

```python
# Global Scope
x = 10 # Global scope
def outer():
    y = 20 # Enclosing scope
    def inner():
        z =30 # Local scope
        res = z + y + x
        print(f"This is Local Scope:{res}")
    res = y + x
    #y = y + x
    inner()
    print(f"This is Enclosing scope:{res}")#{y}
outer()
```
