#### Set Methods
- Adding Element To Set
   - add(ele): It adds element to set if element not exists in set,otherwise nothing
   - update(ele): It adds all iterable elements in sets
- Remove elements from set
   - remove(ele): It removes the element from set if element exists in set oterwise 'KeyError'
    - discard(ele): It removes the element from set if element exists in set otherwise nothing
    - pop(): It remove and return the element from set
- Set operations
 - union(anthor_sets); It returns the all common elements from all sets
 - intersection(anthor_sets):It returns the common elements from the all sets
 - difference(anthor_sets): It returns the set elements those are not lines in anthor_sets
 - symmetric_difference(anthor_sets): It returns the non-common elements from all sets

```python
# Adding elements from set
s = {1,2,3,4,5,7}
s.add(6)
print(s)
s.update([10,20,30])
print(s)
s.add(-1)
print(s)
s.add(1) # 1 doesn't add to set because set contain 1
print(s)
```

```python
# remove elements from set
s = {2,3,5,6,7,-1}
s.remove(7)
print(s)
s.remove(1)# KeyError
print(s)
s.discard(1)
print(s)
```

```python
# Set operations
s1 = {1,2,3,4,5}
s2 = {2,5,7,3,1}
s3 = {9,8}
print(s1.union(s2))
print(s1.union(s2,s3))
```

```python
# Set operations with their symbols
s1 = {1,2,3,4,5}
s2 = {2,5,7,3,1}
s3 = {9,8}
print(s1 | s2)
print(s1 | s2 |s3)
```

```python
s1 = {1,2,3,4,5}
s2 = {3,4,5,6,7}
print(s1.intersection(s2))
```

```python
# Set operations with their symbols
s1 = {1,2,3,4,5}
s2 = {3,4,5,6,7}
print(s1 & s2)
```

```python
s1 = {1,2,3}
s2 = {2,5,7}
print(s1.difference(s2))
```

```python
# Set operations with their symbols
s1 = {1,2,3}
s2 = {2,5,7}
print(s1 - s2)
```

```python
s1 = {1,2,3,4,5}
s2 = {2,5,7,3,1}
print(s1.symmetric_difference(s2))
```
