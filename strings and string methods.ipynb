#### String Methods
-Case Conversion Methods
 -upper() : It converts entire string into uppercase
 -lower() : It converts entire string into lowercase
 -title() : It converts every word first character into uppercase and remining into lowercase
 -swapcase() :It converts lower letter into upper and upper letter into lower
 -capitalize() :It converts first character of string into uppercase and remaining into lower

```python
string="The ocean is blue and sky is BLUE"
lower=string.lower()
print(lower)
print(string.upper())
print(string.title())
print(string.swapcase())
print( string.capitalize())
```

```python
string="The ocean is blue and sky is BLUE12345@"
lower=string.lower()
print(lower)
print(string.upper())
print(string.title())
print(string.swapcase())
print( string.capitalize())
```

```python
# white space remove methods
# strip():It removes the white space to both sides of string
#Lstrip(): It removes the lift side space of a string / It removes the Leading white space
# rstrip(): It removes the right side space of a string /It removes the Lagging white space
string="          Codegnan  "
print(string)
print(string.strip())
print(string.lstrip())
print(string.rstrip())
```

```python
# white space remove methods
# strip():It removes the white space to both sides of string
#Lstrip(): It removes the lift side space of a string / It removes the Leading white space
# rstrip(): It removes the right side space of a string /It removes the Lagging white space
string="@@@@@@@@@@Code  @  gnan@@@@@@@@  "
print(string)
print(string.strip("@"))
print(string.lstrip("@"))
print(string.rstrip())
```

```python
# split():It return the list of substring and by default splits at white space
string ="The ocean is blue and sky is BLUE 123442@"
print(string.split())
print(string.split('is'))
print(string.split('Z'))
print(string.split('blue'))
print(string.split('@'))
```

```python
# Indexing finding Methods
# index(sub-string):It returns the first occurance index of substring if exists otherwise"ValueError"
# find(sub-string)
string ="The ocean is blue and sky is blue"
print(string.index("o"))
print(string.index('is'))
print(string.index('is',14)) # It starts checking from 14th index
print(string.index('Z'))
```

```python
# Indexing finding Methods
# index(sub-string):
# find(sub-string)
string ="The ocean is blue and sky is blue"
print(string.find("o"))
print(string.find('is'))
print(string.find('is',14)) # It starts checking from 14th index
print(string.find('Z'))
```

```python
# replace(old,new):It replace the old string with new_string in string
string ="The ocean is blue and sky is blue"
print(string.replace('blue','Red'))
print(string.replace('blue','Red',1)) # It replaces only one occurences
```

```python
# "step".join(iterable):It joins the all elements with seperator
lst =['The', 'ocean', 'is', 'blue', 'and', 'sky', 'is', 'BLUE']
print("".join(lst))
print(" ".join(lst))
print("_".join(lst))
```

```python
#### strin checking methods
-islower():It returns True entire string is in lowercase
-isupper():It returns True entire string is in uppercase
-istitle():It returns True when string is titlr format
-isalpha():It returns True when string contains Alphebets(i.e lower and uppercase)
-isalnum():It returns True when string either Alphebet or digits
-isdigit():It reteurns True when string contain only digits
-startswith(substring): it returns True when string startswith the sub_string
-endswith(substring):it returns True when string endswith the sub_string
```

```python
print("hello".islower())
print('Hello'.islower())
print('HELLO'.isupper())
print('1234'.isdigit())
print('abcACB'.isalpha())
print('1234'.isalnum())
print('1234abc'.isalnum())
print('HelloWorld'.startswith('H'))
print('HelloWorld'.endswith('H'))
```
