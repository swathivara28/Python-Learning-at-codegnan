```python
# reading integer from user
# integer to float, string, bool
n = int(input())
print(type(n),n)
print(float(n))
print(str(n))
print(bool(n))
```

```python
# reading Float value from user
# float to int, string, bool
val = float(input())
print(type(val),val)
print(str(val))
print(bool(val))
print(int(val))
```

```python
# reading complex value
real = int(input())
img = int(input())
c = complex(real,img)
print(type(c),c)
```

```python
# reading string value
s = input() # "123"
print(type(s),s)
print(int(s))
print(float(s))
print(list(s)) # It returns list of characters
print(tuple(s))  # It returns tuple of characters
print(set(s))   # It returns set of characters
```

```python
# Read list of string
l = list(map(str,input().split()))
print(l)
```

```python
# Reading Tuple of integers
t = tuple(map(int,input().split()))
print(list(t))
print(set(t))
print(str(t))
```

```python
# Reading set of integers
s = set(map(int,input().split()))
print(list(s))
print(tuple(s))
print(str(s))
```
